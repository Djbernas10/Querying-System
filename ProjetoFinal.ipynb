Ao correr o código, a lista <b>results</b> irá guardar o conteudo das páginas da wikipedia pesquisadas (que têm o seu nome dentro da lista de Queries)

```python
# Importar módulo da api da wikipedia e efetuar searcch com as linguagens de programação
# Por fim guardar resultados
import wikipediaapi
wiki_search = wikipediaapi.Wikipedia(
        language='pt',
        extract_format=wikipediaapi.ExtractFormat.WIKI
)
Queries=['Python','C#','C','Kotlin','Ruby','R','Swift','Java','C++','Javascript','PHP','Html']
results=[]
for query in Queries:
    wiki_results = wiki_search.page(query)
    results.append(wiki_results.text)
results
```

O código seguinta irá normalizar na sua totalidade os documentos guardas através da pesquisa pela API do código anterior, fazendo uso da biblioteca criada por nós (previamente instlada no inicio do documento).
Neste caso iremos retirar os acentuação de todas as palavras, retirar as marcas de html, retirar a pontuação, retirar as stopwords e por fim colocar o texto na sua totalidade em minúsculas.

```python
# Importar módulo de tratamento de texto efetuado no trabalho anterior
from Text import text as t
accented = [t.remove_accented_chars(text) for text in results] # Remover acentos
special_removal = [t.remove_special_characters(text) for text in accented] # Remover caracteres especiais
stopwords_removal = [t.stopwords_removal(text,language = "portuguese") for text in special_removal] # Remover stopwords
newlines = [t.remove_extra_newlines(text) for text in stopwords_removal] # Remover paragrafos
lowercased = [t.lowerCase(text) for text in newlines] # Colocar em minusculas
lowercased
```

O código seguinte serve para fazer limpeza de caratéres residuais da normalização anterior

```python
# Efetuar replace de -,/,/br,// por nada 
for i, element in enumerate(lowercased):
    element = element.replace('-','')
    element = element.replace('/br','')
    element = element.replace('/','')
    element = element.replace('//','')
    results[i] = element
```

Neste caso mandamos para dentro da função a quantidade total de ficheiros pesquisado incialmente, e a lista de tokens retirados anteriromente dos documentos. Depois, procedemos ao processo de stemming do termo em questão e verificamos se no nosso dicionario de posicioes já existe esse termo para o documento em especifico.


Caso exista esse termo no documento adicionamos o Offset (<i>numero que indica a posição relativa da palavra face às palavras totais de um documento</i>) desse termo à lista de offsets presente no documento em especifico, caso esse Offset nao se verifique, criamos uma entrada nova para esse termo no documento.

<div class="toc"><ul class="toc-item"><li><span><a href="#Tarefa-3:-Free Text Query" data-toc-modified-id="-Criação do index posiocional"><span class="toc-item-num">4&nbsp;&nbsp;</span>Tarefa 4: phrase query</a></span></li></ul></div>

Por fim executamos um código que irá devolver os documentos onde a query usada para pesquisa aparece de modo absoluto (os documentos devolvidos têm as palavras pesquisadas com a ordem absoluta de como foram pesquisadas).
Para isso faremos uso do package previamente instalado de queries, onde no mesmo estão criadas 4 funções que no final devolvem os docuemtnos onde a frase pesquisada existe.

Na seguinte lista será explicado o que cada função do package de phrase query faz:
<ul>
    <li><b>Intersection_query:</b>
    <p>Interseta as palvras pesquisadas e descobre em que documentos as mesmas aparacem em conjunto (sem prestar atenção se as mesmas aparcem seguidas)</p>
    </li>
    <br>
    <li><b>wordIds:</b>
    <p>Cria um dicionario local onde guarda os Offsets das palavras da query e os associa aos seus documentos</p>
    </li>
    <br>
    <li><b>docOffsets:</b>
    <p>Junta os Offsets de cada documento num dicionario que é separado por documentos.</p>
    </li>
    <br>
    <li><b>subtrOffsets:</b>
    <p>Subtração dos Offsets de cada documento para encontrar efetivamente se as palavras pesquisadas estão de facto seguidas pela ordem pesquisada</p>
    </li>
</ul>

```python
# Pesquisa phrase query
from Text import text
import numpy as np
import nltk
stem = nltk.stem.SnowballStemmer("portuguese")
def intersection_query(ListOfInputTokens):
    keyWords = {} # dicionario de keywords
    # Para cada palavra pesquisada que exista
    # É criado um dicionário com a palavra e o número de documentos onde está presente
    for i in pos_index:
        if i in ListOfInputTokens:
            keyWords[i] = []
            for key, value in pos_index[i][1].items():
                keyWords[i].append(key)
    # Se existirem palavras vamos efetuar a interseção para ver que documentos todas estão presentes
    if keyWords:
        for key,value in keyWords.items():
            keyWords[key] = set(value)  
            res = set.intersection(*keyWords.values())
    else:
        res="Vazio"
    return res

def wordIds(ListOfInputTokens,intersected):
    ids = {}
    k=0
    for palavra in ListOfInputTokens: # para cada palavra na query
        if palavra not in pos_index:
            break
        else:
            ids[palavra]={} # criar uma entrada por palavra
            if k >= len(ListOfInputTokens):
                pass
            else:
                for key, value in pos_index[palavra][1].items() :
                    if key in intersected:
                        ids[palavra][key] = value # adiciona 
        k+=1
    
    return ids

def docOffsets(posicoes):
    docOffsets = {}
    for i in posicoes: # itera o dicionario de posicoes
        for key,value in posicoes[i].items(): # mostra os offsets de cada palavra do dicionario de pesquisa
            if key not in docOffsets: # se o documento não estiver no arreio
                docOffsets[key] = [] # entao crio o arreio
                docOffsets[key].append(value) # e guardo o arreio de offsets
            else: # caso exista
                docOffsets[key].append(value) #adiciono ao arreio de offsets os offsets da palavra seguint
    
    return docOffsets

def subtrOffsets(DocOff):
    result = {}
    
    for doc,offsets in DocOff.items(): ## entra nos docs 
        subtraction = []
       
        for i in range(len(offsets)): # ve o tamanho do arreio dos offsets
            subtr = np.subtract(DocOff[doc][i], i) # subtrai aos documentos de forma a verificar se são seguidos
            subtraction.append(set(subtr))
            intersected = set.intersection(*subtraction) # compara os valores e faz a intersecao dos offsets

        if len(intersected) > 0:
            result[doc] = intersected
            
    return result

pesquisado = input("Insira a sua pesquisa: ")
## tratamento da query
splitedPesquisado = pesquisado.replace('"','')
accented = t.remove_accented_chars(splitedPesquisado) # Remover acentos
special_removal = t.remove_special_characters(accented)# Remover caracteres especiais
stopwords_removal = t.stopwords_removal(special_removal,language = "portuguese") # Remover stopwords
lowercased = t.lowerCase(stopwords_removal) # colocar em lower case
#criacao da lista de tokens
ListOfInputTokens = text.tokenization(lowercased)
tokens = []

for palavra in ListOfInputTokens:
    tokens.append(stem.stem(palavra))
    
docsIntersected = intersection_query(tokens)
if docsIntersected != "Vazio":
    numbers = list(docsIntersected)
    WordIds = wordIds(tokens,numbers)
    Doff = docOffsets(WordIds)
    result = subtrOffsets(Doff)
    print("A sua pesquisa de forma sequencial está presente nos documentos seguintes:")
    print(list(result.keys()))
else:
    print("Não existem documentos com as"+len(ListOfInputTokens)+" palavras presentes!")
```
